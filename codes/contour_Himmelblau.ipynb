```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def himmelblau(x,y):
    return (x**2+y-11)**2 + (x+y**2-7)**2

x = np.arange(-10,10,0.01)
y = np.arange(-10,10,0.01)
xx,yy = np.meshgrid(x,y)

z = himmelblau(xx,yy)
levels = [0,10,20,40,80,160,320,640,1000,2000,4000,8000,16000,32000,64000,128000,256000,600000]

fig, ax = plt.subplots()
cs = ax.contour(x, y, z, levels, cmap='gist_ncar')
ax.clabel(cs, levels[4::2],  # label every 3 level
          inline=1, fmt='%d', fontsize=8)
ax.set_aspect('equal')
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.xaxis.set_ticks(np.arange(-10,11,5))
ax.yaxis.set_ticks(np.arange(-10,11,5))
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.savefig('himmelblau.pdf', format='PDF')


```
